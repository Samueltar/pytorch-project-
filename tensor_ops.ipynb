```python
import torch

# Contiguous tensor
x = torch.randn(4, 4)
print(x.is_contiguous())  # True
x
y_view = x.view(2, 8)
y_view
y_reshape = x.reshape(2, 8)
y_reshape
# Both are views (share memory), modifications affect original


# Non-contiguous tensor (after transpose)
z = x.transpose(0, 1)
z
print(z.is_contiguous())  # False


# z.view(2, 8)  # Raises RuntimeError

w = z.reshape(2, 8)  # Works! Returns a copy (contiguous tensor)
w
print(w.is_contiguous())  # True
# Modifications to w do NOT affect z
```
